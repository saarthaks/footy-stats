```python
import numpy as np
import pandas as pd
```

```python
cols = ['Rk', 'Player', 'Pos', 'Squad']
df_basic = pd.read_csv('./data/standard_stats.csv', skiprows=1, usecols=cols)

cols = ['Min', 'Gls', 'Ast', 'xG', 'npxG', 'xA']
df_stand = pd.read_csv('./data/standard_stats.csv', skiprows=1, usecols=cols)
for c in cols[1:]:
    df_stand[c] = df_stand[c] / df_stand['Min'] * 90
print(df_stand.columns)
    
cols = ['Sh', 'SoT', 'FK']
df_shoot = pd.read_csv('./data/shooting_stats.csv', skiprows=1, usecols=cols)
for c in cols:
    df_shoot[c] = df_shoot[c] / df_stand['Min'] * 90
print(df_shoot.columns)

cols = ['Cmp', 'Att', 'TotDist', 'PrgDist', 'Cmp.1', 'Att.1', 'Cmp.2', 'Att.2', 
        'Cmp.3', 'Att.3', 'KP', '1/3', 'PPA', 'CrsPA', 'Prog']
df_pass = pd.read_csv('./data/passing_stats.csv', skiprows=1, usecols=cols)
for c in cols:
    df_pass[c] = df_pass[c] / df_stand['Min'] * 90
new_cols = {'Cmp': 'PassCmp', 'Att': 'PassAttm', 'TotDist': 'TotalPassDist', 
            'PrgDist': 'PrgPassDist', 'Cmp.1': 'ShortCmp', 'Att.1': 'ShortAttm', 
            'Cmp.2': 'MedCmp', 'Att.2': 'MedAttm', 'Cmp.3': 'LongCmp', 'Att.3': 'LongAttm',
            'KP': 'KeyPasses', '1/3': 'FinalThirdPass', 'PPA': 'PenPass', 'Prog': 'PrgPassNum'}
df_pass = df_pass.rename(columns=new_cols,)
print(df_pass.columns)

cols = ['Live', 'Dead', 'TB' ,'Press', 'Sw', 'Crs', 'Ground', 'Low', 'High', 'Int']
df_passtype = pd.read_csv('./data/pass_types_stats.csv', skiprows=1, usecols=cols)
for c in cols:
    df_passtype[c] = df_passtype[c] / df_stand['Min'] * 90
new_cols = {'TB': 'ThroughBalls', 'Press': 'PassUnderPress', 'Sw': 'SwitchingPass',
            'Int': 'InterceptedPass'}
df_passtype = df_passtype.rename(columns=new_cols)
print(df_passtype.columns)

cols = ['SCA', 'GCA']
df_goalshot = pd.read_csv('./data/goal_shot_creation_stats.csv', skiprows=1, usecols=cols)
for c in cols:
    df_goalshot[c] = df_goalshot[c] / df_stand['Min'] * 90
print(df_goalshot.columns)

cols = ['Tkl', 'TklW', 'Def 3rd', 'Mid 3rd', 'Att 3rd', 'Tkl.1', 'Att', 'Press', 'Succ',
        'Def 3rd.1', 'Mid 3rd.1', 'Att 3rd.1', 'Blocks', 'Sh', 'Pass', 'Int', 'Clr']
df_def = pd.read_csv('./data/defensive_stats.csv', skiprows=1, usecols=cols)
for c in cols:
    df_def[c] = df_def[c] / df_stand['Min'] * 90
new_cols = {'Def 3rd': 'TklDef3rd', 'Mid 3rd': 'TklMid3rd', 'Att 3rd': 'TklAtt3rd',
            'Tkl.1': 'DribTkl', 'Att': 'DribTklAttm', 'Press': 'NumPressed', 'Succ': 'SuccPress', 
            'Def 3rd.1': 'PressDef3rd', 'Mid 3rd.1': 'PressMid3rd', 'Att 3rd.1': 'PressAtt3rd',
            'Sh': 'ShotsBlocked', 'Pass': 'PassesBlocked', 'Int': 'PassesIntercepted'}
df_def = df_def.rename(columns=new_cols)
print(df_def.columns)
    
cols = ['Def Pen', 'Def 3rd', 'Mid 3rd', 'Att 3rd', 'Att Pen', 'Live', 'Succ', 'Att', 
        '#Pl', 'Megs', 'Carries', 'TotDist', 'PrgDist', 'Targ', 'Rec', 'Miscon', 'Dispos']
df_poss = pd.read_csv('./data/possession_stats.csv', skiprows=1, usecols=cols)
for c in cols:
    df_poss[c] = df_poss[c] / df_stand['Min'] * 90
new_cols = {'Def Pen': 'TouchesDefPen', 'Def 3rd': 'TouchesDef3rd', 'Mid 3rd': 'TouchesMid3rd', 
            'Att 3rd': 'TouchesAtt3rd', 'Att Pen': 'TouchesAttPen', 'Live': 'LiveTouches', 
            'Succ': 'SuccDrib', 'Att': 'AttmDrib', '#Pl': 'PlayersDrib', 'TotDist': 'DistCarried',
            'PrgDist': 'PrgDistCarried', 'Targ': 'TargetOfPass', 'Rec': 'PassesRec'}
df_poss = df_poss.rename(columns=new_cols)
print(df_poss.columns)

cols = ['Fls', 'Recov', 'Won', 'Lost']
df_misc = pd.read_csv('./data/misc_stats.csv', skiprows=1, usecols=cols)
for c in cols:
    df_misc[c] = df_misc[c] / df_stand['Min'] * 90
new_cols = {'Fls': 'Fouls', 'Recov': 'LooseBallsRecov', 'Won': 'AerialsWon', 'Lost': 'AerialsLost'}
df_misc = df_misc.rename(columns=new_cols)
print(df_misc.columns)
```

```python
df = pd.concat([df_stand, df_shoot, df_pass, df_passtype, df_goalshot, df_def, df_poss, df_misc], axis=1)
tmp = df.fillna(0)

valid = df['Min'] > 180
tmp = tmp[valid]
tmp_meta = df_basic[valid]
gk = tmp_meta['Pos'] == 'GK'
df_player = tmp[~gk]
df_meta = tmp_meta[~gk]

player_data = df_player.values
print(player_data.shape)
```

```python
def player_mapping(ranking, name):
    rk2idx = {}
    idx2rk = {}
    for i, d in enumerate(ranking):
        rk2idx[d] = i
        idx2rk[i] = d
    
    name2idx = {}
    idx2name = {}
    for i, d in enumerate(name):
        nm = d[d.index('\\')+1:]
        name2idx[nm] = i
        idx2name[i] = nm
    
    return rk2idx, idx2rk, name2idx, idx2name

rk2idx, idx2rk, name2idx, idx2name = player_mapping(df_meta['Rk'], df_meta['Player'])
```

```python
from sklearn.decomposition import PCA, SparsePCA
import matplotlib.pyplot as plt
```

```python
def similar_players(name, data, normalize=True):
    idx = name2idx[name]
    player = data[idx,:]
    scores = data @ player
    
    if normalize:
        player_norm = np.linalg.norm(player)
        data_norm = np.linalg.norm(data, axis=1)
        scores = scores / data_norm / player_norm
    
    sorted_scores = -np.sort(-scores)[1:]
    ranked_idx = np.argsort(-scores)[1:]
    names = [idx2name[idx] for idx in ranked_idx]
        
    return names, sorted_scores
```

```python
pca = SparsePCA(n_components=16)
pca.fit(player_data[:, 1:])
```

```python
player_trans2 = pca.transform(player_data[:, 1:])
```

```python
names, scores = similar_players('Willian', player_trans2)
for name, score in zip(names, scores):
    print(name, score)
```

```python
pca2 = PCA(n_components=16, whiten=True)
pca2.fit(player_data[:, 1:])
```

```python
plt.figure(figsize=(8,6))
plt.semilogy(pca2.explained_variance_ratio_, '--.')
plt.ylim([1e-10, 1e-1])
plt.grid()
```

```python
player_trans = pca2.transform(player_data[:, 1:])
```

```python
names, scores = similar_players('Richarlison', player_trans)
```

```python
for name, score in zip(names, scores):
    print(name, score)
```
